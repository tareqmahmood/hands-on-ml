```python
import matplotlib.pyplot as plt # plotting
import pandas as pd # data processing
import numpy as np # matrix ops
import sklearn.linear_model # ml
```

```python
# load the data
data = pd.read_csv("lifesat.csv")
```

```python
data
```

```python
x = data["GDP per capita (USD)"]
```

```python
y = data["Life satisfaction"]
```

```python
plt.scatter(x, y)
plt.show()
```

```python
model = sklearn.linear_model.LinearRegression()
```

```python
X = x.to_numpy().reshape(-1, 1)
Y = y.to_numpy().reshape(-1)
```

```python
model.fit(X, Y)
```

```python
x_test = np.array([45000, 50000, 70000, 10000])
x_test.shape
```

```python
x_test = x_test.reshape(-1, 1)
```

```python
y_test = model.predict(x_test)
```

```python
plt.scatter(x, y)
plt.scatter(x_test, y_test, color='red')
plt.show()
```

```python
x_test = np.arange(10000, 70000, 1000)
x_test.shape
```

```python
x_test = x_test.reshape(-1, 1)
```

```python
y_test = model.predict(x_test)
```

```python
plt.scatter(x, y)
plt.plot(x_test, y_test, color='red')
plt.show()
```
