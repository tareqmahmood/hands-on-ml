## Batch Gradient Descent

Batch size: the size of the subset of data to calculate gradient and update weight once

Step: training on one batch

Epoch (Iteration): training on the whole dataset once

## Stochastic Gradient Descent

Batch size = 1

Learning rate: Rule of thumb, start with 0.1, then reduce



Dataset spliiting: 

1. Train set (50000)
2. Validation set (10000)
3. Test set (10000)

```python
X_train, X_valid, X_test, y_train, y_valid, y_test = X[:50000], X[50000:60000], X[60000:], y[:50000], y[50000:60000], y[60000:]
```

```python
n = len(X)
batch_size = 32
```

```python
y_train_5 = (y_train == 5) # True for all 5s, False for all other digits.
y_valid_5 = (y_valid == 5)
y_test_5 = (y_test == 5)
```

```python
from sklearn.linear_model import SGDClassifier
```

```python
for i in range(1, 20):
    sgd_clf = SGDClassifier(random_state=42, max_iter=i)
    sgd_clf.fit(X_train, y_train_5)
    training_score = sgd_clf.score(X_valid, y_valid_5)
    print(f"Step {i}: Training score: {training_score}")
```
